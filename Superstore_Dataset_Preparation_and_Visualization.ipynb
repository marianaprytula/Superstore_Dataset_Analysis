## Visuals

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
def make_autopct(values):
    def custom_autopct(pct):
        total = sum(values)
        count = int(round(pct * total / 100.0))
        return f'{pct:.1f}%\n\n({count})'
    return custom_autopct


explode = (0,  0.1, 0)

fig, ax = plt.subplots(figsize=(8,8))
ax.pie(superstore_df["Segment"].value_counts().values, explode=explode, labels=superstore_df["Segment"].value_counts().index, autopct=(make_autopct(superstore_df["Segment"].value_counts().values)),
       colors = ["violet", "yellow", "green"], shadow=True, startangle=90, textprops={'fontsize': 14})

plt.title('Number of Customers by Segments', fontsize=16)
plt.show()
```

```python
unique_customers_ids = superstore_df.groupby('Segment')['Customer ID'].nunique()


explode = (0,  0.1, 0)

fig, ax = plt.subplots(figsize=(8,8))
ax.pie(unique_customers_ids.values, explode=explode, labels=unique_customers_ids.index, autopct=(make_autopct(unique_customers_ids.values)),
       colors = ["violet", "yellow", "green"], shadow=True, startangle=90, textprops={'fontsize': 14})

plt.title('Number of Distinct Customers by Segments', fontsize=16)
plt.show()
```

```python

fig, ax = plt.subplots(figsize=(10, 6))


bar_labels = superstore_df["Ship Mode"].value_counts().index
bar_colors = ['green', 'yellow', 'blue', 'red']

ax.bar(superstore_df["Ship Mode"].value_counts().index, superstore_df["Ship Mode"].value_counts().values, label=bar_labels, color=bar_colors)

ax.set_ylabel('Number of times used', fontsize=13)
ax.set_title('Shipping Mode specified by the Customer', fontsize=16)
ax.legend(title='Ship Mode Types')

ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)


plt.show()
```

```python
monthly_sales = superstore_df.groupby(superstore_df["Order Date"].dt.to_period("M"))["Sales"].sum().reset_index()
```

```python
monthly_sales
```

```python
monthly_sales['Order Date'] = monthly_sales['Order Date'].dt.to_timestamp()
```

```python

plt.figure(figsize=(30, 10))
plt.plot(monthly_sales['Order Date'], monthly_sales["Sales"], marker='o', linestyle='-', color="pink")

for x, y in zip(monthly_sales['Order Date'], monthly_sales['Sales']):
    plt.annotate(f'{y:.0f}', xy=(x, y), xytext=(0, 10), textcoords="offset points",
                 ha='center', fontsize=10)

plt.title('Total Sales by Months (rounded)', fontsize=16)
plt.xlabel('Year & Month', fontsize=14)
plt.ylabel('Total Sales',  fontsize=14)

date_range = pd.date_range('2014-01', monthly_sales["Order Date"].max(), freq = "MS")
plt.xticks(date_range, rotation=70, fontsize=10)
# plt.grid(True)
plt.show()
```

```python
monthly_profit = superstore_df.groupby(superstore_df["Order Date"].dt.to_period("M"))["Profit"].sum().reset_index()

monthly_profit['Order Date'] = monthly_profit['Order Date'].dt.to_timestamp()
```

```python
plt.figure(figsize=(30, 10))
plt.plot(monthly_profit['Order Date'], monthly_profit["Profit"], marker='o', linestyle='-', color="green")

for x, y in zip(monthly_profit['Order Date'], monthly_profit['Profit']):
    plt.annotate(f'{y:.0f}', xy=(x, y), xytext=(0, 10), textcoords="offset points",
                 ha='center', fontsize=10)

plt.title('Total Profit by Months (rounded)', fontsize=16)
plt.xlabel('Year & Month', fontsize=14)
plt.ylabel('Total Profit',  fontsize=14)

date_range = pd.date_range('2014-01', monthly_profit["Order Date"].max(), freq = "MS")
plt.xticks(date_range, rotation=70, fontsize=10)
# plt.grid(True)
plt.show()
```

```python
monthly_avg_profit_margin = superstore_df.groupby(superstore_df["Order Date"].dt.to_period("M"))["Profit margin"].mean().reset_index()

monthly_avg_profit_margin['Order Date'] = monthly_avg_profit_margin['Order Date'].dt.to_timestamp()
```

```python
plt.figure(figsize=(30, 10))
plt.plot(monthly_avg_profit_margin['Order Date'], monthly_avg_profit_margin["Profit margin"], marker='o', linestyle='-', color="#29b6f6")

for x, y in zip(monthly_avg_profit_margin['Order Date'], monthly_avg_profit_margin['Profit margin']):
    plt.annotate(f'{y:.2f}', xy=(x, y), xytext=(0, 10), textcoords="offset points",
                 ha='center', fontsize=10)

plt.title('Average Profit Margin by Months', fontsize=16)
plt.xlabel('Year & Month', fontsize=14)
plt.ylabel('Avg Profit Margin',  fontsize=14)

date_range = pd.date_range('2014-01', monthly_avg_profit_margin["Order Date"].max(), freq = "MS")
plt.xticks(date_range, rotation=70, fontsize=10)
# plt.grid(True)
plt.show()
```

```python
category_hierarchy = superstore_df.groupby(superstore_df["Category"]).agg(Sub_Category=("Sub-Category",'nunique'),
                                                                           Distinct_Product_number=("Product Name", 'nunique'),
                                                                           Total_Sold_Quantity=("Quantity", 'sum'),
                                                                          AVG_Discount=("Discount", 'mean'),
                                                                          Total_Sales_Value = ("Sales", 'sum'),
                                                                          Total_Profit_Value=("Profit", 'sum')).sort_values(by="Total_Profit_Value", ascending=False).round(2).reset_index()
```

```python
category_hierarchy
```

```python
fig, ax = plt.subplots(figsize=(6, 4.5))

colors =  ['#D5CC48', '#98D548', '#51D548']
cell_colors = [[colors[i]] * len(category_hierarchy.columns) for i in range(len(category_hierarchy))]

ax.axis('off')


table = ax.table(
    cellText=category_hierarchy.values,       # the data values
    colLabels=category_hierarchy.columns,     # column headers
    cellLoc='center',         # align text center
    cellColours=cell_colors,
    loc='center'              # position table in the center
)


table.scale(3.5, 4.5)  # adjust row height
table.auto_set_font_size(False)
table.set_fontsize(12)

plt.title("Categories' Info", fontsize=16)
plt.show()
```

```python
customer_profit = superstore_df.groupby(superstore_df["Customer ID"]).agg(Sales=('Sales', 'sum'),
                                                                          Profit=('Profit', 'sum')).sort_values(by='Profit', ascending=False).round(0).astype(int).reset_index()
```

```python
top_10_customer_profit = customer_profit.head(10)
top_10_customer_profit
```

```python
customers = top_10_customer_profit['Customer ID']
sales = top_10_customer_profit['Sales']
profits = top_10_customer_profit['Profit']
x = np.arange(len(customers))  # positions
width = 0.4

fig, ax = plt.subplots(figsize=(15, 7))

bars1 = ax.bar(x - width/2, sales, width, label='Total Sales', color='tab:blue')
bars2 = ax.bar(x + width/2, profits, width, label='Total Profit', color='tab:green')

for bar in bars1:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height + 4, f'{height}', ha='center', va='bottom', fontsize=11)

for bar in bars2:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height + 4, f'{height}', ha='center', va='bottom', fontsize=11)

ax.set_xlabel('Customer ID', fontsize=11)
ax.set_ylabel('Sales / Profit',fontsize=11)
ax.set_title('Total Sales and Profit per Customer', fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels(customers)
ax.legend(fontsize=11)
plt.tight_layout()
plt.show()
```
